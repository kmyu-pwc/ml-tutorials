Axes위에 여러개의 차트 그리기

```python
x = np.arange(0, 5, 0.2) # 0 ~5 사이에 0.2 단위로 실수 생성
y1 = x
y2 = x**2
y3 = x**3

fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(x, y1, 'r--', label= 'y1')
ax.plot(x, y2, 'bs', label= 'y2')
ax.plot(x, y3, 'g^', label= 'y3')

plt.legend()# 범례 표시
plt.show()
```

```python
# 다른 방법

fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(x, y1, 'r--', x, y2, 'bs', x, y3, 'g^')

plt.legend(['y1', 'y2', 'y3'], loc = 'best') # loc = `best` 적당한 위치에 배치(loc : location)
plt.show()
```

- ``plt.ylabel()`` → ``ax.set_ylabel()``

- ``plt.ylim()`` → ``ax.set_ylim()``

산점도 속성 수정하기

```python
class_label = {'Iris-setosa': 1, 'Iris-versicolor':2, 'Iris-virginica':3}
color_map = ["#00798c", "#d1495b", '#edae49', '#66a182']

fig, ax = plt.subplots(figsize = (5, 5))

ax.scatter(iris.loc[iris['Species']=='Iris-setosa', 'SepalLength'], iris.loc[iris['Species']=='Iris-setosa', 'SepalWidth'],
           c=color_map[0], marker = 'o', s=40, edgecolor = 'black',label=class_label['Iris-setosa'])
            
ax.scatter(iris.loc[iris['Species']=='Iris-versicolor', 'SepalLength'], iris.loc[iris['Species']=='Iris-versicolor', 'SepalWidth'],
           c=color_map[1], marker = 's', s=40, edgecolor = 'black', label=class_label['Iris-versicolor'])
            
ax.scatter(iris.loc[iris['Species']=='Iris-virginica', 'SepalLength'], iris.loc[iris['Species']=='Iris-virginica', 'SepalWidth'],
           c=color_map[2], marker = '^', s=40, edgecolor = 'black',label=class_label['Iris-virginica'])

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
plt.legend(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.show()
```

for loop를 사용하려 그리기

```python
class_label = {'Iris-setosa': 1, 'Iris-versicolor':2, 'Iris-virginica':3}
color_map = ["#00798c", "#d1495b", '#edae49', '#66a182']
marker = ['o', 's', '^']

fig, ax = plt.subplots(figsize = (5, 5))

for i, target in enumerate(class_label):
    ax.scatter(iris.loc[iris['Species']== target, 'SepalLength'], iris.loc[iris['Species']==target, 'SepalWidth'],
              c=color_map[i], marker=marker[i], s = 40, edgecolors='black', label=class_label[target])
    
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
plt.legend(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.show()
```

```python
class_label = {'Iris-setosa': 1, 'Iris-versicolor':2, 'Iris-virginica':3}
color_map = ["#00798c", "#d1495b", '#edae49', '#66a182']
marker = ['o', 's', '^']

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 5))

for i, target in enumerate(class_label):
    ax1.scatter(iris.loc[iris['Species']== target, 'SepalLength'], iris.loc[iris['Species']==target, 'SepalWidth'],
              c=color_map[i], marker=marker[i], s = 40, edgecolors='black', label=class_label[target])
    
ax1.set_xlabel('Sepal Length')
ax1.set_ylabel('Sepal Width')

ax1.legend(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])

for i, target in enumerate(class_label):
    ax2.scatter(iris.loc[iris['Species']== target, 'PetalLength'], iris.loc[iris['Species']==target, 'PetalWidth'],
              c=color_map[i], marker=marker[i], s = 40, edgecolors='black', label=class_label[target])
    
ax2.set_xlabel('Petal Length')
ax2.set_ylabel('Petal Width')

ax2.legend(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.show()
```

**참고 : KDE(Kernel Density Estimation, 확률밀도추정) plot(중요)**  
- histogram은 구간(bins) 설정에 따라 다른 결과와 해석을 내릴수 있는 우려가 있다.  
- 그 대안으로 KDE가 사용됨 (데이터 분포를 smoothing한 곡선으로 표현)

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (12, 4))

cancer.loc[cancer['diagnosis']=='M', 'radius_mean'].plot(kind = 'hist', ax = ax1, bins=30, density = True, # 빈도 대신 밀도로 
                                                          color = color_map[0], alpha = 0.6, label='malignant')

cancer.loc[cancer['diagnosis']=='B', 'radius_mean'].plot(kind = 'hist', ax = ax1, bins=30, density = True,
                                                          color = color_map[1], alpha = 0.6, label='benign')

cancer.loc[cancer['diagnosis']=='M', 'radius_mean'].plot(kind = 'kde', ax = ax1, style = 'b--', linewidth = 1)
cancer.loc[cancer['diagnosis']=='B', 'radius_mean'].plot(kind = 'kde', ax = ax1, style = 'r--', linewidth = 1)

ax1.set_xlabel('radius_mean')
ax1.legend(['malignant', 'benign'])

cancer.loc[cancer['diagnosis']=='M', 'area_mean'].plot(kind = 'hist', ax = ax2, bins=30, density = True,
                                                          color = color_map[0], alpha = 0.6, label='malignant')
cancer.loc[cancer['diagnosis']=='B', 'area_mean'].plot(kind = 'hist', ax = ax2, bins=30, density = True,
                                                          color = color_map[1], alpha = 0.6, label='benign')
cancer.loc[cancer['diagnosis']=='M', 'area_mean'].plot(kind = 'kde', ax = ax2, style = 'b--', linewidth = 1)
cancer.loc[cancer['diagnosis']=='B', 'area_mean'].plot(kind = 'kde', ax = ax2, style = 'r--', linewidth = 1)


ax2.set_xlabel('aream_mean')
ax2.legend(['malignant', 'benign'])

plt.show()
```

#### 2.4 box plot  
- 데이터의 5가지 요약 통계 정보를 차트로 표현  
- 5가지 요약통계정보  
    - min,Q1(median에서 하위 25%지점), median(중앙값, Q2, 50%지점), Q3(median에서 상위 25% 지점), max

<img src="img/ch04/03_box_plot.jpg" width="500">

- whis : 일반적으로 1.5
- min : Q1 - whis*(Q3-Q1)  
- max : Q3 + whis*(Q3-Q1)  
- min ~ max를 이탈하는 값을 이상치(outlier)로 간주함

```python
fig, ax = plt.subplots(figsize = (12, 5))
cancer.boxplot(column='radius_mean', ax = ax)
ax.set_title('boxplot radius_mean')
plt.show()
```

- outlier가 동그라미로 표시됨

```python
fig, ax = plt.subplots(figsize = (12, 5))
ax.boxplot([cancer[cancer['diagnosis']=='B']['radius_mean'].values,
       cancer[cancer['diagnosis']=='M']['radius_mean'].values], labels=['B', 'M'], showmeans=True)
ax.set_title('radius_mean by diagnosis')
plt.show()
```

DataFrame을 사용한 방법

```python
fig, ax = plt.subplots(figsize = (12, 5))
cancer.boxplot(column='radius_mean', by = 'diagnosis', ax = ax)
ax.set_title('radius_mean by diagnosis')
plt.show()
```

속성 조정하기

```python
color_pal = ['#708090', '#be9b7b', '#feb2a8',
	'#58668b',  '#4f372d', '#ECD59F', '#ABD1DC', '#7E9680', 
    	'#E69D45', '#D4CFC9', '#E79796', '#C6C09C', '#dec3c3',
    	'#4b86b4', '#f9caa7', '#d2d4dc', '#ebf4f6', '#29a8ab' ]


fig, ax = plt.subplots(figsize = (12, 5))
data['energy'].sort_values(ascending = False).plot(kind = 'bar', ax = ax, fontsize = 12 , color= color_pal, 
                                                   edgecolor='#404040', rot= 45) # 큰순서대로 정렬

for p in ax.patches:
    ax.text(p.get_x()+p.get_width()/2., p.get_height(), 
            str(p.get_height()), ha='center', fontsize = 13) # 값 표시

plt.show()
```
